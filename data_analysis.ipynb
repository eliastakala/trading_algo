```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#visualize close and adjusted close prices
plt.figure(figsize=(16,8))
plt.title('Tesla')
plt.xlabel('Days')
plt.ylabel('Close Price USD ($)')
plt.plot(df['Close TSLA'])
plt.plot(df['Adj Close TSLA'])
plt.legend(['Close', 'Adj Close'], loc='upper left')
```
